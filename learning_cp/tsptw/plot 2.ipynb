```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import os
import re

sns.set(rc={"figure.figsize":(14, 10)})
```

```python
os.getcwd()
```

```python
path = "exp_2022-04-28T16-42-24/"

dir_list = os.listdir(path)
training = dict()
evaluation = dict()
for file in dir_list:
    m = re.search('.*_(.*).csv', file)
    if m:
        name = m.group(1)
        if name.startswith("training"):
            training[name] = pd.read_csv(path + file)
        else:
            evaluation[name] = pd.read_csv(path + file)

training_only_0 = dict()
evaluation_only_0 = dict()

for key, df in training.items():
    df["Heuristic"] = key
    training_only_0[key] = df[df["Solution"] == 0]

for key, df in evaluation.items():
    df["Heuristic"] = key
    evaluation_only_0[key] = df[df["Solution"] == 0]

total_evaluation_only_0 = pd.concat([df for df in evaluation_only_0.values()], axis=0, ignore_index=True)
total_trained_only_0 = total_evaluation_only_0[total_evaluation_only_0["Heuristic"].str.startswith("trained")]
total_training_only_0 = pd.concat([df for df in training_only_0.values()], axis=0, ignore_index=True)
```

```python
# training_only_0["training"].describe()
%matplotlib inline
```

```python
sns.set(rc={"figure.figsize":(24, 16)})
fig, axs = plt.subplots(nrows=2, ncols=3)
sns.regplot(x= "Episode", y ="Nodes",
           data=total_evaluation_only_0[total_evaluation_only_0["Heuristic"] == "trained"], fit_reg=True, x_estimator=np.mean, ax=axs[0][0])
sns.regplot(data=total_evaluation_only_0[total_evaluation_only_0["Heuristic"] == "trained"], y="Reward", x="Episode", color="green", logx=True, ax=axs[0][1], x_estimator=np.mean)
sns.scatterplot(data=total_evaluation_only_0[total_evaluation_only_0["Heuristic"] == "trained"], y="Loss", x="Episode", ax=axs[0][2], color="orange")
sns.regplot(data=total_evaluation_only_0[total_evaluation_only_0["Heuristic"] == "trained"], y="Time", x="Episode", color="brown", logx=True, ax=axs[1][0], x_estimator=np.mean)
sns.regplot(data=training_only_0["training"], y="Reward", x="Episode", scatter_kws={'alpha':0.3}, logx=True, ax=axs[1][1], color="green")
sns.scatterplot(data=training_only_0["training"], y="Loss", x="Episode", ax=axs[1][2], alpha=0.5, color="orange")
```
